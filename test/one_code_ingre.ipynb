```python
import pandas as pd
import numpy as np
```

```python
ingredientes = pd.read_csv('ingredientes.csv')
```

```python
ingredientes = ingredientes.drop(['strDrink','strIngredient12','strIngredient13','strIngredient14','strIngredient15'],axis=1)
```

```python
columnas = ingredientes.columns.values
columnas
```

```python
for i in columnas:
    if i != 'idDrink':
        ingredientes[i] = ingredientes[i].str.lower()
```

```python
lista = ['idDrink']
for i in columnas:
    if i != 'idDrink':
        lista.extend(ingredientes[i].unique())   
```

```python
ingredientes_unicos = list(set(lista))
```

```python
len(ingredientes_unicos)
```

```python
ing_one_code = pd.DataFrame(columns=ingredientes_unicos)
```

```python
primera_col = ing_one_code.pop('idDrink')
                               
ing_one_code.insert(0 ,'idDrink',primera_col)                        
```

```python
ing_one_code.idDrink = ingredientes.idDrink
```

```python
for i in range(len(ingredientes)):
    drink_i = ingredientes.iloc[i]
    for j in  drink_i:
        if isinstance(j,str):
            idDrink = ingredientes.iloc[i,0]
            ing_one_code.loc[ing_one_code['idDrink']==idDrink,j] = 1


   
```

```python
ing_one_code = ing_one_code.replace(np.nan,0)
```
