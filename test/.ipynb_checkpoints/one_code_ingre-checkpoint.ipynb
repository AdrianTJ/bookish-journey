```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Read file as csv
ingredientes = pd.read_csv('ingredientes.csv')
```

```python
# Drop empty columns including strDrink (the name of the cocktail) and columns strIngridient 12 - strIngridient 15
# There is no cocktail with more than 12 ingredients.
ingredientes = ingredientes.drop(['strDrink','strIngredient12','strIngredient13','strIngredient14','strIngredient15'],axis=1)
```

```python
# Generate an array with the name of the columns

columnas = ingredientes.columns.values
columnas
```

```python
# Change all ingredients to lowercase letters
for i in columnas:
    if i != 'idDrink':
        ingredientes[i] = ingredientes[i].str.lower()
```

```python
# Put in a list all ingredients appending all columns.

lista = ['idDrink']
for i in columnas:
    if i != 'idDrink':
        lista.extend(ingredientes[i])
        
len(lista)
```

```python
# Selecting unique ingredients of this list.
ingredientes_unicos = list(set(lista))
```

```python
# Gives the number of ingredientes contained in the list.
len(ingredientes_unicos)
```

```python
# Put all unique ingredients into a pandas dataframe as columns
ing_one_code = pd.DataFrame(columns=ingredientes_unicos)
ing_one_code
```

```python
# Extract idDrink from the columns
primera_col = ing_one_code.pop('idDrink')

# Insert in position "0" the labek 'idDrink'                               
ing_one_code.insert(0 ,'idDrink',primera_col)      
```

```python
for i in range(len(ingredientes)):
    drink_i = ingredientes.iloc[i]
    for j in  drink_i:
        if isinstance(j,str):
            idDrink = ingredientes.iloc[i,0]
            ing_one_code.loc[ing_one_code['idDrink']==idDrink,j] = 1


```

```python
# Replace all NaN with "0"
ing_one_code = ing_one_code.replace(np.nan,0)
```
